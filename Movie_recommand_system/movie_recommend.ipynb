```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
# Merge two datasets for the usefull datasets. 
movies = movies.merge(credits,on='title')
```

```python
# select some usefull columes for more proccesing.
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# importing ast(abstract syntax tree) model for the literal_eval() function
import ast
```

```python
# ast.literal_eval() => convert the string into list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# select name word value from the genres colume.
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
# select name word value from the keywords colume.
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# this function is written for to find the first three name of cast and converted into a list.
def convert_3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
```

```python
movies.head()
```

```python
# This function is written for to find the Director name from the crew.
def find_Director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(find_Director)
```

```python
movies.head()
```

```python
# for converting the string para into list for mearging the all column
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# the following function is writen for to remove the space from the words.
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Concat the column into given column is tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# Creating new data set for more oprations.
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
# converting the list into string.
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
# converting in lower case.
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
# we use the countvectorizer model for to convert the text into vector array.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
# nural lang toolkit for the below models
import nltk
```

```python
# we used the porterstemmer for removing the similar words.
# stemming is the process for removing the suffixes from words
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
# Example text:['played','play','playing']  output: play
stem('played')
# EX-2 text:['replacing','replacement','replace'] output: replac
stem('replacement')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
# used cosine_similarity for to find out the similarity between two vectors.
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
# This function recommend the movie to the user.
def recommend(movie):
    
    #this for finding title index and name of the movie.
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    #this for the finding similarity of that movie at index.
    distances = similarity[movie_index]
    
    #this for sorted the macting movie to the user movie.
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]
    
    # this loop for give the movie title to the user that macthed.
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
